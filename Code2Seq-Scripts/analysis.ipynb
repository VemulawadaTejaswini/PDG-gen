# Count total number of generated Java files

```python
import os

def find_java_files(directory):
  """Recursively finds all Java files in a directory and its subdirectories.

  Args:
    directory: The directory to search.

  Returns:
    A list of all Java files found.
  """

  java_files = []
  for root, directories, filenames in os.walk(directory):
    for filename in filenames:
      if filename.endswith(".java"):
        java_files.append(os.path.join(root, filename))

  return java_files
```

# Count total number of generated PDG files

```python
import os

def find_java_files(directory):
  """Recursively finds all Java files in a directory and its subdirectories.

  Args:
    directory: The directory to search.

  Returns:
    A list of all Java files found.
  """

  java_files = []
  for root, directories, filenames in os.walk(directory):
    for filename in filenames:
      if filename.endswith(".txt"):
        file_path = os.path.join(root, filename)
        file_pt = open(file_path, "r")
        file_content = file_pt.readlines()
        if len(file_content) >= 5:
          java_files.append(os.path.join(root, filename))
        file_pt.close()

  return java_files

directory = "/home/tomy495/CS21MTECH12001/Research/API-misuse/Code2Seq-Data/java-large/processed-pdg-data"
java_files = find_java_files(directory)

print(len(java_files))

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
```
